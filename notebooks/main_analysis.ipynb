```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
column_names = ['Frequency', 'Angle of Attack', 'Chord Length', 'Free-stream Velocity', 'Suction Side Displacement', 'Sound Pressure Level']
data = pd.read_csv('../data/airfoil_self_noise.dat', sep='\t', header=None, names=column_names)

print(data.head())


# Some histogram charting
plt.figure(figsize=(6, 3))
data['Angle of Attack'].plot(kind='hist')
plt.title('Angle of Attack distribution')
plt.xlabel('Angle of Attack')
plt.ylabel('Frequency')
plt.show()

# Some Scatter plotting
plt.figure(figsize=(6, 4))
plt.scatter(data['Chord Length'], data['Angle of Attack'], marker='o', linestyle='-', color="red", alpha=0.7)
plt.title('AoA vs Chord Length')
plt.xlabel('Chord length')
plt.ylabel('Angle of Attack ')
plt.show()


print(data.info())
print(data.describe())

data['Free-stream Velocity'].value_counts()
```
